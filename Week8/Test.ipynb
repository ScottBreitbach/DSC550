## Step 12: Adjust skewed data (fare)
If you go back and look at the histograms of Fare, you’ll see that it is very skewed…many low cost fares, not very many high cost fares.  Log Transformation is a good method to use on highly skewed data.  


##### Check the distribution with a histogram:

```python
# Set up the figure size
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# Create a histogram
plt.hist(df['Fare'], bins=40)
plt.xlabel('Fare', fontsize=20)
plt.ylabel('Counts', fontsize=20)
plt.tick_params(axis='both', labelsize=15)
plt.show()
```

##### Create a new column from the log transformation of the **Fare** column:

```python
# Load libraries
import numpy as np
```

```python
# Perform a natural log-transformation
def log_transformation(data):
    return data.apply(np.log1p)

df['Fare_log1p'] = log_transformation(df['Fare'])
```

## Step 15: Evaluation Metrics
Remember, we are trying to predict if a passenger has survived or not so this is a classification problem.  
There are many algorithms that could be used but we’re going to use logistic regression.  

```python
# Load libraries
from sklearn.linear_model import LogisticRegression

from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.classifier import ClassificationReport
from yellowbrick.classifier import ROCAUC
```

### Confusion Matrix

```python
# Instantiate the classification model 
model = LogisticRegression(multi_class="auto", solver='liblinear') # penalty='l1'
```

```python
# The ConfusionMatrix visualizer taxes a model
classes = ['Not_survived', 'Survived']
cm = ConfusionMatrix(model, classes=classes, percent=False)

# Fit fits the passed model. This is unnecessary if you pass the visualizer a pre-fitted model
cm.fit(X_train, y_train)

# To create the ConfusionMatrix, we need some test data. Score runs predict() 
# on the data and then creates the confusion_matrix from scikit learn.
cm.score(X_val, y_val)

# Change fontsize of the labels in the figure
for label in cm.ax.texts:
    label.set_size(20)

# How did we do?
cm.poof()
```

```python
correct = 158 + 68
incorrect = 24 + 18
total = correct + incorrect

print(f"Overall percent predicted correctly: {round(correct/total*100, 2)}")
```

### Precision, Recall, & F1 Score

```python
# Precision, Recall, and F1 Score
# set the size of the figure and the font size 
#%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 7)
plt.rcParams['font.size'] = 20

# Instantiate the visualizer
visualizer = ClassificationReport(model, classes=classes)

visualizer.fit(X_train, y_train)  # Fit the training data to the visualizer
visualizer.score(X_val, y_val)  # Evaluate the model on the test data
g = visualizer.poof()
```

* **Precision**: the ability to not label an instance positive that is actually negative
* **Recall**: the ability to find all positive instances
* **F1 Score**: a weighted harmonic mean of *precision* and *recall*  

Scores looked good across the board for both classes, though prediction was consistently better for `Not_survived` than for `Survived`.
