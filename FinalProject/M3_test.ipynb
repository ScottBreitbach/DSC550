### Think About the Data

* The **Class** variable will be the *target* and all other variables will be the features.
* All of the variables are **numeric** with the exception of the target variable, which is **categorical**.
* The *target* variable contains **7** unique categories, but there are *not* an equal number of each.
* Some of the beans appear to be much larger than others and should be relatively easy to classify, while others could prove more difficult.
* **None** of the columns appear to contain **NaNs**.

Some of the features are very strongly correlated with each other, indicating that they might be good candidates for dimensionality reduction because they might not be adding any additional information (i.e. they are redundant).

#### Find Highly Correlated Features

```python
## LOOK AT HIGHLY CORRELATED FEATURES
# (i.e. likely to be redundant)

# Create features and target
features = beanDF.select_dtypes(include='number')
target = beanDF.Class

# Create correlation matrix
corr_matrix = features.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),
                                 k=1).astype(np.bool))

# Find index of feature columns with high correlation
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Features to drop
print("Features with high correlation:\n", to_drop)
```

Of these 7 variables that are considered redundant (to be removed), 5 of them are in the list of 6 variables to keep from the previous step. I'll definitely need to adjust some thresholds so that I keep more than 1 feature variable.

```python
def makeHistogram(var, df=beanDF):
    # Create a histogram
    plt.hist(df[var], bins=40)
    plt.xlabel(var, fontsize=20)
    plt.ylabel('Counts', fontsize=20)
    plt.tick_params(axis='both', labelsize=15)
    plt.show()
```

```python
def compareHist(var, suffix, df=beanDF):
    '''Compares, side-by-side, the original 
    histogram with the transformed histogram.'''
    # Make subplots
    fig, axes = plt.subplots(nrows=1, ncols=2)

    # Specify the features of interest
    num_features = [var, var+suffix]
    xAxes = num_features
    yAxes = ['Counts'] * 2

    # Draw histrograms
    axes = axes.ravel()
    for idx, ax in enumerate(axes):
        ax.hist(df[num_features[idx]].dropna(), bins=40)
        ax.set_xlabel(xAxes[idx], fontsize=20)
        ax.set_ylabel(yAxes[idx], fontsize=20)
        ax.tick_params(axis='both', labelsize=15)

    # plt.suptitle("Histograms of Seed Attributes", fontsize=25)

    plt.show()
```

#### Log(10) Transformations

```python
# Get a list of the columns for log-transformation
cols_to_tranform = ['Area', 'Perimeter', 'MajorAxisLength', 
                    'MinorAxisLength', 'ConvexArea', 'EquivDiameter']
```

```python
# Perform a log10-transformation
def log_10_transform(data):
    return data.apply(np.log10)
```

```python
# Add columns of log-transformed data
for item in cols_to_tranform:
    beanDF[item+'_log10'] = log_10_transform(beanDF[item])
```

```python
# Show comparitive histograms for transformed data
for item in cols_to_tranform:
    compareHist(item, '_log10', beanDF)
```

I tried both natural log and log10. The histograms generated by both looked pretty much the same, so I went with log10 because it's more intuitive to read.

#### Cube Transformations

```python
# Perform a cubed transformation
def cube_transform(data):
    return data**3
```

```python
# Show comparitive histograms for transformed data
for item in cols_to_tranform:
    compareHist(item, '_cubed', beanDF)
```

#### Other Transformations

```python
# Load libraries
from sklearn import preprocessing
import numpy as np
```

```python
# Get a list of the columns for transformation
cols_to_tranform = ['Solidity', 'ShapeFactor4']
```

```python
# Standardize column 0-1 and transform to 5th exponent
def data_transform(data):
    feature = np.array(data).reshape(-1,1)
    minmax_scale = preprocessing.MinMaxScaler(feature_range=(0, 1))
    scaled_feature = minmax_scale.fit_transform(feature)
    return scaled_feature**5
```

```python
# Add columns of transformed data
for item in cols_to_tranform:
    beanDF[item+'_xformed'] = data_transform(beanDF[item])
```

```python
# Show comparitive histograms for transformed data
for item in cols_to_tranform:
    compareHist(item, '_xformed', beanDF)
```
