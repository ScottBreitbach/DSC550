# 1.2 Exercise: Working with Arrays

### 2D Arrays

1) Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive).  

```python
# nrows = 3
# ncols = 2
nrows = 10
ncols = 50
size = nrows * ncols
```

```python
array = np.random.randint(1, 501, size).reshape(nrows, ncols)
array
```

Load it to a DataFrame

```python
df = pd.DataFrame(array)
df
```

2) Calculate the sum of each row

```python
# print("Sum of Row:")
# df.sum(axis=1)
```

```python
print("Sum of each row:")
np.sum(array, axis=1)
```

The sum of each column

```python
# print("Sum of Column:")
# df.sum()
```

```python
print("Sum of each column:")
np.sum(array, axis=0)
```

The sum of all entries

```python
print("Sum of entire matrix:")
np.sum(array)
```

3) Determine the minimum of all entries

```python
print("Smallest number in the matrix:")
np.min(array)
```

Maximum

```python
print("Largest number in the matrix:")
np.max(array)
```

Average

```python
print("Average of all entries:")
np.mean(array).round(2)
```

Standard deviation

```python
np.std(array).round(2)
```
